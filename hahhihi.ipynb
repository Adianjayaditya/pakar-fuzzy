```python
import pandas as pd

# Baca file Excel dan pilih sheet yang akan diproses
file_path = 'hasil hasil belajar.xlsx'
sheet1 = pd.read_excel(file_path, sheet_name='Sheet1')
sheet2 = pd.read_excel(file_path, sheet_name='Sheet2')
rule = pd.read_excel(file_path, sheet_name='Sheet3')


def convert_to_seconds(time_str):
    try:
        parts = time_str.split(', ')
        minutes = int(parts[0].replace(' menit', ''))
        seconds = int(parts[1].replace(' detik', ''))
        return minutes * 60 + seconds
    except:
        return None

sheet1['lama_berlatih (min 5, mak 15)'] = sheet1['lama_berlatih (min 5, mak 15)'].apply(convert_to_seconds)
sheet2['lama_berlatih (min 5, mak 15)'] = sheet2['lama_berlatih (min 5, mak 15)'].apply(convert_to_seconds)

sheet1 = sheet1.drop(columns=["no"], axis=1)
sheet2 = sheet2.drop(columns=["no"], axis=1)
rule = rule.drop(columns=["Unnamed: 0"], axis=1)
```

```python
def ubah_kepercayaan(row):
    if row["lulus"] == "lulus":
        return row["kepercayaan"]
    else:
        return 100 - row["kepercayaan"]
new_rule = rule.copy()
new_rule["kepercayaan"] = new_rule.apply(ubah_kepercayaan, axis=1)

# Tampilkan hasil
new_rule
```

```python
# Parameter batas nilai
batas1 = 25
batas2 = 75
batas3 = 100

def kategorikan_kepercayaan(nilai, batas1, batas2, batas3):
    if 0 <= nilai < batas1:
        return "kecil"
    elif batas1 <= nilai < batas2:
        return "sedang"
    elif batas2 <= nilai <= batas3:
        return "besar"
    else:
        return "tidak valid"

new_rule["kepercayaan"] = new_rule["kepercayaan"].apply(kategorikan_kepercayaan, args=(batas1, batas2, batas3))
new_rule
```

```python
# Fungsi keanggotaan segitiga
def keanggotaan_segitiga(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif x == b:
        return 1
    elif a < x < b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)

# Fungsi keanggotaan trapesium
def keanggotaan_trapesium(x, a, b, c, d):
    if x <= a or x >= d:
        return 0
    elif b <= x <= c:
        return 1
    elif a < x < b:
        return (x - a) / (b - a)
    elif c < x < d:
        return (d - x) / (d - c)
```

```python
# PARAMETER WAKTU BELAJAR
wb_kecil1 = 0
wb_kecil2 = 0
wb_kecil3 = 356
wb_kecil4 = 454
wb_sedang1 = 356
wb_sedang2 = 454
wb_sedang3 = 527
wb_besar1 = 454
wb_besar2 = 527
wb_besar3 = 3600
wb_besar4 = 3600

# PARAMETER LAMA LATIHAN
lb_kecil1 = 0
lb_kecil2 = 0
lb_kecil3 = 367
lb_kecil4 = 621
lb_sedang1 = 367
lb_sedang2 = 621
lb_sedang3 = 1056
lb_besar1 = 621
lb_besar2 = 1056
lb_besar3 = 9000
lb_besar4 = 9000

# PARAMETER NILAI
nl_kecil1 = 0
nl_kecil2 = 0
nl_kecil3 = 66.7
nl_kecil4 = 83
nl_sedang1 = 66.7
nl_sedang2 = 83
nl_sedang3 = 100
nl_besar1 = 83
nl_besar2 = 100
nl_besar3 = 100

# PARAMETER KELULUSAN
kl_kecil1 = 0
kl_kecil2 = 0
kl_kecil3 = 66.7
kl_kecil4 = 83
kl_sedang1 = 66.7
kl_sedang2 = 83
kl_sedang3 = 100
kl_besar1 = 83
kl_besar2 = 100
kl_besar3 = 100

# INPUT
waktu_belajar = 517.0	
lama_latihan = 1226.0	
nilai = 99.0


def fuzzyfy(wb, lb, nl):
    wb_fuzzy = {
    'kecil': keanggotaan_trapesium(wb, wb_kecil1, wb_kecil2, wb_kecil3, wb_kecil4), 
    'sedang': keanggotaan_segitiga(wb, wb_sedang1, wb_sedang2, wb_sedang3), 
    'besar': keanggotaan_trapesium(wb, wb_besar1, wb_besar2, wb_besar3, wb_besar4)
    }
    lb_fuzzy = {
    'kecil': keanggotaan_trapesium(lb, lb_kecil1, lb_kecil2, lb_kecil3, lb_kecil4), 
    'sedang': keanggotaan_segitiga(lb, lb_sedang1, lb_sedang2, lb_sedang3), 
    'besar': keanggotaan_trapesium(lb, lb_besar1, lb_besar2, lb_besar3, lb_besar4)
    }
    nl_fuzzy = {
    'kecil': keanggotaan_trapesium(nl, nl_kecil1, nl_kecil2, nl_kecil3, nl_kecil4), 
    'sedang': keanggotaan_segitiga(nl, nl_sedang1, nl_sedang2, nl_sedang3), 
    'besar': keanggotaan_segitiga(nl, nl_besar1, nl_besar2, nl_besar3)
    }
    return wb_fuzzy, lb_fuzzy, nl_fuzzy

kelulusan_fuzzy = {
    'kecil': keanggotaan_trapesium(nilai, kl_kecil1, kl_kecil2, kl_kecil3, kl_kecil4), 
    'sedang': keanggotaan_segitiga(nilai, kl_sedang1, kl_sedang2, kl_sedang3), 
    'besar': keanggotaan_segitiga(nilai, kl_besar1, kl_besar2, kl_besar3)
    }

wb_fuzzy, lb_fuzzy, nl_fuzzy = fuzzyfy(waktu_belajar, lama_latihan, nilai)
```

```python
def inferensi_mamdani(x, y, z, rules):
    hasil_inferensi = []  # Untuk menyimpan hasil dari setiap rule yang digunakan
    for i, rule in rules.iterrows():
        # Ambil nilai keanggotaan dari input berdasarkan rule
        waktu_belajar = x[rule["rata2 Lama akses materi"]]
        lama_jawab_soal = y[rule["lama_berlatih (min 5, mak 15)"]]
        nilai = z[rule["nilai"]]
        # Hitung derajat keanggotaan berdasarkan operator AND (min)
        min_value = min(waktu_belajar, lama_jawab_soal, nilai)
        
        if min_value > 0:  # Rule ini digunakan jika min_value > 0
            hasil_inferensi.append({"rule": i + 1, "kelulusan": rule["kepercayaan"], "value": min_value})
            rule_text = (
                    f"IF waktu belajar {rule['rata2 Lama akses materi']} "
                    f"AND lama jawab soal {rule['lama_berlatih (min 5, mak 15)']} "
                    f"AND nilai {rule['nilai']} "
                    f"THEN kelulusan {rule['kepercayaan']} "
                    f"({min_value:.3f})"
                )
            print(rule_text)
    return hasil_inferensi

hasil_inferensi = inferensi_mamdani(wb_fuzzy, lb_fuzzy, nl_fuzzy, new_rule)

# Tampilkan hasil inferensi
# print("Rules yang hit:")
# for hasil in hasil_inferensi:
#     print(f"Rule {hasil['rule']} -> Kelulusan: {hasil['kelulusan']}, Nilai: {hasil['value']}")
```

```python
def agregasi(hasil_inferensi):
    """
    hasil_inferensi: Hasil inferensi berupa daftar rule yang hit.
    Menghasilkan nilai maksimum untuk setiap himpunan fuzzy output dan indeks rule yang menghasilkan nilai maksimum.
    """
    kelulusan_fuzzy = {"kecil": 0, "sedang": 0, "besar": 0}
    max_rule_indices = {"kecil": None, "sedang": None, "besar": None}
    
    for hasil in hasil_inferensi:
        kelulusan = hasil["kelulusan"]
        value = hasil["value"]
        rule_index = hasil["rule"]
        
        # Perbarui nilai maksimum dan simpan indeks rule jika lebih besar
        if value > kelulusan_fuzzy[kelulusan]:
            kelulusan_fuzzy[kelulusan] = value
            max_rule_indices[kelulusan] = rule_index
    
    return kelulusan_fuzzy, max_rule_indices


kelulusan_fuzzy, max_rule_indices = agregasi(hasil_inferensi)

# Cetak hasil
print(f"Kelulusan fuzzy: {kelulusan_fuzzy}")
print(f"Indeks rule maksimum: {max_rule_indices}")
```

```python
def defuzzifikasi(hasil_agregasi, max_rule_indicex):
    numerator = 0
    denominator = 0
    for kelulusan, value in hasil_agregasi.items():
        if kelulusan == "kecil" and max_rule_indices["kecil"] != None:
            cf = rule.iloc[max_rule_indices["kecil"] - 1]["kepercayaan"] / 100
            central_area = (kl_kecil4 - kl_kecil1)/2 + kl_kecil1
            numerator = numerator + kelulusan_fuzzy["kecil"] * cf * central_area
        elif kelulusan == "sedang" and max_rule_indices["sedang"] != None:
            cf = rule.iloc[max_rule_indices["sedang"] - 1]["kepercayaan"] / 100
            central_area = (kl_sedang3 - kl_sedang1)/2 + kl_sedang1
            numerator = numerator + kelulusan_fuzzy["sedang"] * cf * central_area
        elif kelulusan == "besar" and max_rule_indices["besar"] != None:
            cf = rule.iloc[max_rule_indices["besar"] - 1]["kepercayaan"] / 100
            central_area = (kl_besar3 - kl_sedang1)/2 + kl_besar1
            numerator = numerator + kelulusan_fuzzy["besar"] * cf * central_area
        denominator += value
    return numerator / denominator

hasil = defuzzifikasi(kelulusan_fuzzy, max_rule_indices)
hasil
```
